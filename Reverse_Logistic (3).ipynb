```python
from sklearn.preprocessing import LabelEncoder
ln = LabelEncoder()
df['Type'] = ln.fit_transform(df['Type'])
df['Delivery Status'] = ln.fit_transform(df['Delivery Status'])
df['Customer Segment'] = ln.fit_transform(df['Customer Segment'])
df['Department Name'] = ln.fit_transform(df['Department Name'])
df['Market'] = ln.fit_transform(df['Market'])
df['Order Status'] = ln.fit_transform(df['Order Status'])
```

```python
df['Order Status'].value_counts()
# Here class SUSPECTED_FRAUD denote as 8
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop(labels=['Order Status'], axis=1),df['Order Status'], test_size=0.2,random_state=0)
```

```python
print( x_train.shape)
print( y_train.shape)
print( x_test.shape)
print( y_test.shape)
```

```python
import xgboost as xgb
classifier = xgb.XGBClassifier()
```

```python
classifier.fit(x_train, y_train)
```

```python
prediction = classifier.predict(x_test)
y_train_predict = classifier.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, f1_score, recall_score, roc_auc_score
```

```python
print("Training Accuracy", accuracy_score(y_train, y_train_predict))
print("Testing Accuracy", accuracy_score(y_test, prediction))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, prediction))
pd.crosstab(y_test, prediction)
```
